```python
import os
import pandas as pd
import cv2

# Paths
csv_path = '/content/LicensePlateDetection/train_annotations.csv'
image_folder = '/content/LicensePlateDetection/images/train'
label_folder = '/content/LicensePlateDetection/labels/train'

# Create label folder if it doesn't exist
os.makedirs(label_folder, exist_ok=True)

# Load CSV
df = pd.read_csv(csv_path)

# Group annotations by image
grouped = df.groupby('img_id')

# Process each image and write YOLO-format annotations
for image_name, group in grouped:
    image_path = os.path.join(image_folder, image_name)

    if not os.path.exists(image_path):
        print(f"Image not found: {image_path}")
        continue

    img = cv2.imread(image_path)
    h, w = img.shape[:2]

    label_path = os.path.join(label_folder, os.path.splitext(image_name)[0] + '.txt')
    with open(label_path, 'w') as f:
        for _, row in group.iterrows():
            # Convert to YOLO format
            x_center = ((row['xmin'] + row['xmax']) / 2) / w
            y_center = ((row['ymin'] + row['ymax']) / 2) / h
            width = (row['xmax'] - row['xmin']) / w
            height = (row['ymax'] - row['ymin']) / h

            # Write: class_id (0 for license plate), x_center, y_center, width, height
            f.write(f"0 {x_center:.6f} {y_center:.6f} {width:.6f} {height:.6f}\n")

print("✅ Annotations converted to YOLO format and saved.")
```

```python
import os
from PIL import Image
import matplotlib.pyplot as plt
import easyocr

# Initialize OCR reader
reader = easyocr.Reader(['en'])

# Folder path containing cropped images
folder_path = "/content/LicensePlateDetection/cropped_images"

display_limit = 30
display_count = 0

# Loop through each image in the folder
for image_name in sorted(os.listdir(folder_path)):
    if not image_name.lower().endswith(('.png', '.jpg', '.jpeg')):
        continue

    image_path = os.path.join(folder_path, image_name)
    result = reader.readtext(image_path)

    print(f"\nImage path: {image_path}")

    if not result:
        print("No text detected.")
    else:
        print("OCR Output:")
        for _, text, _ in result:
            print(f"Detected Text: {text}")

    # Show only the first 30 images
    if display_count < display_limit:
        img = Image.open(image_path)
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"OCR Preview: {image_name}")
        plt.show()
        display_count += 1
```

```python
from ultralytics import YOLO
import os
import cv2
import easyocr
import pandas as pd

# Load model and OCR
model = YOLO("/content/runs/detect/train/weights/best.pt")
reader = easyocr.Reader(['en'], gpu=True)

# Test images folder
image_folder = "/content/LicensePlateDetection/Test_images"

# Data to collect results
submission_data = []

# Process each image
for image_name in sorted(os.listdir(image_folder)):
    image_path = os.path.join(image_folder, image_name)
    image = cv2.imread(image_path)

    if image is None:
        print(f"Skipping unreadable image: {image_name}")
        continue

    results = model(image)[0]
    print(f"\nImage: {image_name}")

    digit_vector = [0] * 10
    found_text = ""

    if results.boxes:
        for i, box in enumerate(results.boxes):
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            cropped = image[y1:y2, x1:x2]

            # OCR
            ocr_result = reader.readtext(cropped)
            if ocr_result:
                for _, text, conf in ocr_result:
                    found_text += text

    else:
        print("No license plate detected.")

    # Only keep digits and update digit vector
    for ch in found_text:
        if ch.isdigit():
            digit_vector[int(ch)] = 1

    print(f"Detected Text: {found_text}")
    print(f"Digit Vector: {digit_vector}")

    # Append row: [image_id, digit0, digit1, ..., digit9]
    image_id = os.path.splitext(image_name)[0]
    submission_data.append([image_id] + digit_vector)

# Create submission DataFrame
columns = ["Id"] + [str(i) for i in range(10)]
submission_df = pd.DataFrame(submission_data, columns=columns)

# Save as CSV
submission_df.to_csv("submission.csv", index=False)
print("\n✅ Submission file saved as submission.csv")
```
